```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns


names = ['Sequence Name', 'mcg', 'gvh', 'alm', 'mit', 'vac', 'nuc', 'class']
dataset = pd.read_csv("output.csv")
print(dataset.head())


def euclidean_distance(data1, data2):
    distance = 0
    for i in range(len(data1)):
        distance += (data1[i] - data2[i]) ** 2
    return distance ** (1/2)

def get_neighbors(train_data, train_labels, test_data, k):
    distances = []
    for i in range(len(train_data)):
        dist = euclidean_distance(test_data, train_data[i])
        distances.append((train_data[i], train_labels[i], dist))
    distances.sort(key=lambda x: x[2])
    neighbors = distances[:k]
    return neighbors

def predict_classification(neighbors):
    class_votes = {}
    for neighbor in neighbors:
        label = neighbor[1]
        if label in class_votes:
            class_votes[label] += 1
        else:
            class_votes[label] = 1
    sorted_votes = sorted(class_votes.items(), key=lambda x: x[1], reverse=True)
    return sorted_votes[0][0]

def k_nearest_neighbors(X_train, Y_train, X_test, k):
    predictions = []
    for i in range(len(X_test)):
        neighbors = get_neighbors(X_train, Y_train, X_test[i], k)
        prediction = predict_classification(neighbors)
        predictions.append(prediction)
    return predictions


X = dataset.iloc[:, 1:7].values
Y = dataset.iloc[:, 7].values

train_ratio = 0.8
validation_ratio = 0.2

total_samples = len(X)
train_samples = int(total_samples * train_ratio)
validation_samples = total_samples - train_samples

indices = list(range(total_samples))
random.shuffle(indices)

train_indices = indices[:train_samples]
validation_indices = indices[train_samples:]

X_train = X[train_indices]
Y_train = Y[train_indices]
X_validation = X[validation_indices]
Y_validation = Y[validation_indices]
```

```python
k = int(input("Enter the value of K: "))
```

```python
predictions = k_nearest_neighbors(X_train.tolist(), Y_train.tolist(), X_validation.tolist(), k)


correct = sum(1 for p, actual in zip(predictions, Y_validation) if p == actual)
accuracy = correct / float(len(Y_validation))

pred_train = k_nearest_neighbors(X_train.tolist(), Y_train.tolist(), X_train.tolist(), k)
correct_train = sum(1 for pred, actual in zip(pred_train, Y_train) if pred == actual)
acc_train = correct_train / float(len(Y_train))
print("KNN Accuracy: ", accuracy)
print("KNN Train Accuracy: ",acc_train)

if not isinstance(Y_train[0], int):
    class_mapping = {label: idx for idx, label in enumerate(set(Y_train))}
    Y_train = [class_mapping[label] for label in Y_train]

feature_names = ['mcg', 'gvh', 'alm', 'mit', 'vac', 'nuc']
data_for_scatter = pd.DataFrame(X_train, columns=feature_names)

scatter_matrix = pd.plotting.scatter_matrix(data_for_scatter, c=Y_train, marker='o', s=40, hist_kwds={'bins': 15}, figsize=(12, 12), cmap='rainbow')

for ax in scatter_matrix.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize=12)
    ax.set_ylabel(ax.get_ylabel(), fontsize=12)

classes = list(set(Y_train))
colors = plt.cm.rainbow([c / len(classes) for c in range(len(classes))])
class_labels = [f'Class {c}' for c in classes]
legend_labels = dict(zip(class_labels, colors))

plt.subplots_adjust(right=0.85) 
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=label, markersize=10, markerfacecolor=color) for label, color in legend_labels.items()]
plt.legend(handles=legend_handles, title="Classes", title_fontsize=12, fontsize=10, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
import numpy as np

def plot_decision_boundary(X, Y, k, feature1, feature2):
    # Create a mesh grid for the feature space
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    # Predict the class labels for each point in the mesh grid
    Z = np.array(k_nearest_neighbors(X, Y, np.c_[xx.ravel(), yy.ravel()].tolist(), k))
    Z = Z.reshape(xx.shape)

    # Plot the decision boundaries
    plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
    plt.xlabel(feature1)
    plt.ylabel(feature2)

# Set the features 
feature1 = 'mcg'
feature2 = 'gvh'

# Visualize scatter plot for the selected features
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data_for_scatter, x=feature1, y=feature2, hue=Y_train, palette='rainbow', markers='o')
plt.title(f'Scatter Plot of {feature1} vs {feature2} with Class Labels')
plt.legend(title="Classes", title_fontsize=12)
plt.grid(True)

# Set the limits for x and y axes to zoom in
plt.xlim(X_train[:, 0].min() - 0.1, X_train[:, 0].max() + 0.1)
plt.ylim(X_train[:, 1].min() - 0.1, X_train[:, 1].max() + 0.1)

# Plot the decision boundaries for the selected features
plot_decision_boundary(X_train, Y_train, k, feature1, feature2)

plt.show()
```
